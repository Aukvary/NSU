```python
# 4
"""Напишите программу, которая считывает целое число и выводит соответствующую ему римскую цифру. Если число находится вне диапазона 
[1;10], то программа должна вывести текст «ошибка» (без кавычек)."""
```

```python
num = int(input())
if 1 <= num <= 3: print("I" * num)
elif num == 4: print("IV")
elif 5 <= num <= 8: print("V" + "I" * (num - 5))
elif num == 9: print("IX")
elif num == 10: print("X")
else: print("ошибка")
```

```python
# 5
"""Напишите программу, которая принимает на вход число и в зависимости от условий выводит текст «YES» (без кавычек) либо «NO» (без кавычек).
Условия:
если число нечётное, то вывести «YES»;
если число чётное в диапазоне от 2 до 5 (включительно), то вывести «NO»;
если число чётное в диапазоне от 6 до 20 (включительно), то вывести «YES»;
если число чётное и больше 20, то вывести «NO»."""
```

```python
num = int(input())

if (num % 2 != 0): print("YES")
elif 2 <= num <= 5: print("NO")
elif 6 <= num <= 20: print("YES")
elif num > 20: print("NO")
```

```python
# 6
"""Даны две различные клетки шахматной доски. Напишите программу, которая определяет, может ли слон попасть с первой клетки на вторую одним ходом. 
Программа получает на вход четыре числа от 1 до 8 каждое, задающие номер столбца и номер строки сначала для первой клетки, потом для второй клетки.
Программа должна вывести «YES», если из первой клетки ходом слона можно попасть во вторую, или «NO» в противном случае."""
```

```python
# 7
"""Даны две различные клетки шахматной доски. Напишите программу, которая определяет, может ли конь попасть с первой клетки на вторую одним ходом.
Программа получает на вход четыре числа от 1 до 8 каждое, задающие номер столбца и номер строки сначала для первой клетки, потом для второй клетки.
Программа должна вывести «YES», если из первой клетки ходом коня можно попасть во вторую, или «NO» в противном случае."""
```

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())

dx = abs(x1 - x2)
dy = abs(y1 - y2)

if dx == 1 and dy == 2: 
    print("YES")
elif dx == 2 and dy == 1:
    print("YES")
else:
    print("NO")
```

```python
# 8
"""Даны две различные клетки шахматной доски. Напишите программу, которая определяет, может ли ферзь попасть с первой клетки на вторую одним ходом. 
Программа получает на вход четыре числа от 1 до 8 каждое, задающие номер столбца и номер строки сначала для первой клетки, потом для второй клетки. 
Программа должна вывести «YES», если из первой клетки ходом ферзя можно попасть во вторую, или «NO» в противном случае."""
```

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())

dx = abs(x1 - x2)
dy = abs(y1 - y2)

if dx == dy: 
    print("YES")
elif dx == 0 or dy == 0: 
    print("YES")
else: 
    print("NO")
```

```python
# 9
"""Саша, засевший за компьютером и не ведущий активный образ жизни, «немного» поднабрал в весе. 
Осталось всего 60 дней до лета, а хочется быть в форме. Вот Саша и решился на похудение. Все дни до лета он пронумеровал от 
1 до 60 (включительно). Перед началом похудения у Саши был вес 100 кг, а своей целью он поставил достичь веса 
88 кг (или меньше). Он решил худеть на одну и ту же массу ежедневно.

Напишите программу, которая принимает на вход текущий день и текущий вес Саши. Программа должна вывести фразу:

«Все идет по плану» (без кавычек), если Саша удаётся держать планку в похудении и его вес ниже либо равен тому, 
который он запланировал на текущий день «Что-то пошло не так» (без кавычек), если Саша не очень старается и его вес выше того, 
который он запланировал на текущий день. Также программа должна вывести информацию о номере дня похудения, текущем весе Саши и цели по
весу на текущий день в формате:
<номер дня> ДЕНЬ: ТЕКУЩИЙ ВЕС = <текущий вес Саши> кг, ЦЕЛЬ по ВЕСУ = <цель по весу на текущий день> кг

На вход программе подаются два числа (каждое на отдельной строке): номер дня похудения (целое число) и текущий вес Саши (действительное число).
Программа должна вывести текст в соответствии с условием задачи.
В 1-й день похудения Саша уже должен похудеть!!!

Input 1:
34
93.1
Output 1:
Все идет по плану
#34 ДЕНЬ: ТЕКУЩИЙ ВЕС = 93.1 кг, ЦЕЛЬ по ВЕСУ = 93.2 кг

Input 2:
1
99.9
Output 2:
Что-то пошло не так
#1 ДЕНЬ: ТЕКУЩИЙ ВЕС = 99.9 кг, ЦЕЛЬ по ВЕСУ = 99.8 кг

Input 3:
53
89
Output 3:
Все идет по плану
#53 ДЕНЬ: ТЕКУЩИЙ ВЕС = 89.0 кг, ЦЕЛЬ по ВЕСУ = 89.4 кг               """
```

```python
day = int(input())
current = float(input())
need = 100 - day * (12 / 60)

print("Все идет по плану" if current <= need else "Что-то пошло не так")
print(f'#{day} ДЕНЬ: ТЕКУЩИЙ ВЕС = {current} кг, ЦЕЛЬ по ВЕСУ = {need} кг')
```
