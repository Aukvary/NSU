```python
# Исправьте код
def get_max_index(numbers):
    max_index = 0
    max_value = numbers[0] 
    for index, value in enumerate(numbers): 
        if max_value < value: 
            max_index = index
            max_value = value
    return max_index
```

```python
#6
"""В НГУ имя студента считается корректным, если оно начинается с заглавной латинской буквы,
за которой следуют строчные латинские буквы. 
Например, имена Sasha и Li считаются корректными, а имена vasya, Yo17, AleskandRRR нет. 
Также у каждого студента имеется идентификационный номер, представленный натуральным числом, который выдается при поступлении в университет. 
К примеру, если в университете обучается 1000 студентов, то новый прибывший студент получит идентификационный номер равный 1001.

Реализуйте функцию get_id(), которая принимает два аргумента:
    names — список имен студентов, обучающихся в НГУ
    name — имя поступающего студента
Функция должна возвращать идентификационный номер, который получит поступающий в НГУ студент, при этом
если имя студента name не является строкой (тип str), функция должна возбуждать исключение:
-TypeError('Имя не является строкой')                
если имя студента name является строкой (тип str), но не представляет собой корректное имя, функция должна возбуждать исключение:
-ValueError('Имя не является корректным')

Input 1:
names = ['Sasha', 'Vasya', 'Dima']
name = 'Egor'
print(get_id(names, name))
Output 1:
4

Input 2:
names = ['Sasha', 'Vasya', 'Dima', 'Arthur']
name = 'Ruslan1337'
try:
    print(get_id(names, name))
except ValueError as e:
    print(e)
Output 2:
Имя не является корректным

Input 3:
names = ['Sasha', 'Vasya', 'Dima', 'Arthur', 'Ruslan']
name = ['E', 'd', 'u', 'a', 'r', 'd']
try:
    print(get_id(names, name))
except TypeError as e:
    print(e)
Output 3:
Имя не является строкой       """
```

```python
def get_id(names, name):
    if type(name) != str:
        raise TypeError('Имя не является строкой')
    
    if name != name.capitalize():
        raise ValueError('Имя не является корректным')
    
    if any(c.isdigit() for c in name):
        raise ValueError('Имя не является корректным')
    
    names.append(name)
    return len(names)


names = ['Sasha', 'Vasya', 'Dima']
name = 'Egor'
print(get_id(names, name))

names = ['Sasha', 'Vasya', 'Dima', 'Arthur']
name = 'Ruslan1337'
try:
    print(get_id(names, name))
except ValueError as e:
    print(e)

names = ['Sasha', 'Vasya', 'Dima', 'Arthur', 'Ruslan']
name = ['E', 'd', 'u', 'a', 'r', 'd']
try:
    print(get_id(names, name))
except TypeError as e:
    print(e)
```

```python
#7
"""Назовем пароль хорошим, если:
    его длина равна 9 или более символам
    в нем присутствуют большие и маленькие буквы любого алфавита
    в нем имеется хотя бы одна цифра
Напишите программу, которая требует ввода нового пароля до тех пор, пока не будет введен хороший.
На вход программе подается произвольное количество паролей, каждый на отдельной строке. Гарантируется, что среди них присутствует хороший.
Для каждого введенного пароля программа должна вывести текст:
    LengthError, если длина введенного пароля меньше 9 символов
    LetterError, если в нем все буквы имеют одинаковый регистр либо отсутствуют
    DigitError, если в нем нет ни одной цифры
    Success!, если введенный пароль хороший
После ввода хорошего пароля все последующие пароли должны игнорироваться.
Приоритет вывода сообщений об ошибке в случае невыполнения нескольких условий: LengthError, затем LetterError, а уже после DigitError.

Input 1:
arr1
Arrrrrrrrrrr
arrrrrrrrrrrrrrr1
Arrrrrrr1
Output 1:
LengthError
DigitError
LetterError
Success!

Sample Input 2:
battery
Battery123
Battery2022
Battery2023
Battery2024
Output 2:
LengthError
Success!         """
```

```python
while 1:
    password = input()

    try:
        if len(password) < 9: 
            raise Exception('LenghtError')
        if not any(c.isupper() for c in password):
            raise Exception('LetterError')
        if not any(c.islower() for c in password):
            raise Exception('LetterError')
        if all(not c.isdigit() for c in password):
            raise Exception('DigitError')
        
        print('Success!')
        break

    except Exception as ex: print(ex) 
```

```python
#13
"""Реализуйте функцию range_sum() с использованием рекурсии, которая принимает три аргумента в следующем порядке:

numbers — список целых чисел
start — целое неотрицательное число
end — целое неотрицательное число
Функция должна суммировать все числа из списка numbers от numbers[start] до numbers[end] включительно и возвращать полученный результат.
Рассмотрим первый тест. Диапазону индексов [3;7] в переданном списке принадлежат числа ​4,5,6,7,8​, сумма которых равна:
4+5+6+7+8=30
Гарантируется, что start <= end.

Input 1:
print(range_sum([1, 2, 3, 4, 5, 6, 7, 8, 9], 3, 7))
Output 1:
30

Input 2:
print(range_sum([1, 2, 3, 4, 5, 6, 7, 8, 9], 0, 8))
Output 2:
45

Input 3:
print(range_sum([1, 2, 3, 4, 5, 6, 7, 8, 9], 0, 0))
Output 3:
1                """
```

```python
def range_sum(nums, start, end):
    if start > end: return 0
    if start == end: return nums[end]

    return nums[start] + nums[end] + range_sum(nums, start + 1, end - 1)

print(range_sum([1, 2, 3, 4, 5, 6, 7, 8, 9], 3, 7))
```

```python
#14
"""Возводить в степень можно гораздо быстрее, чем за n умножений. Для этого нужно воспользоваться следующими рекуррентными соотношениями:
a**n = (a**2)**(n/2), при четном n
a**n = a * a**(n-1), при нечетном n

Реализуйте функцию get_fast_pow() с использованием рекурсии, которая принимает два аргумента в следующем порядке:
a — положительное целое число
n — неотрицательное целое число
Функция должна вычислять значение a в степени n, используя алгоритм быстрого возведения в степень, и возвращать полученный результат.
При решении не используйте оператор возведения в степень **.
Input 1:
print(get_fast_pow(2, 10))
Output 1:
1024

Input 2:
print(get_fast_pow(5, 2))
Output 2:
25

Input 3:
print(get_fast_pow(2, 100))
Output 3:
1267650600228229401496703205376 """
```

```python
def get_fast_pow(a, n):
    if not n: return 1

    if n % 2 == 0:
        return get_fast_pow(a*a, n // 2)
    else:
        return a * get_fast_pow(a, n - 1)
    
print(get_fast_pow(int(input()), int(input())))
```

```python
#14
"""Реализуйте функцию is_power() с использованием рекурсии, которая принимает один аргумент:
number — натуральное число
Функция должна возвращать значение True, если number является степенью числа 2, или False в противном случае.
Input 1:
print(is_power(512))
Output 1:
True

Input 2:
print(is_power(15))
Output 2:
False

Input 3:
print(is_power(1))
Output 3:
True                 """
```

```python
def is_power(n):
    if n == 1: return False
    if n == 2: return True
    return (n % 2 == 0) and is_power(n // 2)


for i in range(1, 33):
    print(f'{i}: {is_power(i)}')
```

```python
#15
"""Реализуйте функцию is_palindrome() с использованием рекурсии, которая принимает один аргумент:
string — произвольная строка
Функция должна возвращать значение True, если переданная строка является палиндромом, или False в противном случае.
Палиндром — текст, одинаково читающийся в обоих направлениях.
Пустая строка является палиндромом, как и строка, состоящая из одного символа.

Input 1:
print(is_palindrome('stepik'))
Output 1:
False

Input 2:
print(is_palindrome('level'))
Output 2:
True

Input 3:
print(is_palindrome('122333221'))
Output 3:
True                  """
```

```python
def is_palindrome(s):
    if not s: return True
    return s[0] == s[-1] and is_palindrome(s[1:-1])

print(is_palindrome('stepik'))

print(is_palindrome('level'))

print(is_palindrome('122333221'))
```
