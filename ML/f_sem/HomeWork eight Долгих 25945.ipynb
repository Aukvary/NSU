```python
#1 - семантика
"""Реализуйте функцию find_most_similar(target_word, words_dict, top_k=3), которая находит top_k самых семантически близких слов
к целевому слову на основе их векторных представлений.
Функция должна:
    Использовать косинусную близость
    Исключать целевое слово из результатов
    Возвращать отсортированный список кортежей (слово, схожесть)
    Обрабатывать случай отсутствия целевого слова в словаре

Используйте любой словарь
dct1 = {'кошка': [0.9, 0.1, 0.8, 0.2],
           'собака': [0.8, 0.2, 0.7, 0.3],
           'тигр': [0.7, 0.3, 0.9, 0.1],
           'лев': [0.6, 0.4, 0.8, 0.2],
           'рыба': [0.1, 0.9, 0.2, 0.8], 
           'птица': [0.2, 0.8, 0.3, 0.7] }

dct2 = {'яблоко': [0.9, 0.1, 0.2],
          'груша': [0.8, 0.2, 0.3],
          'банан': [0.7, 0.3, 0.4],
          'апельсин': [0.6, 0.4, 0.1],
          'машина': [0.1, 0.9, 0.8], 
          'дом': [0.2, 0.8, 0.7]}

dct3 = {'красный': [0.9, 0.1, 0.2],
          'оранжевый': [0.8, 0.3, 0.2],
          'желтый': [0.7, 0.5, 0.1],
          'зеленый': [0.1, 0.9, 0.2],
          'синий': [0.2, 0.3, 0.9],
          'фиолетовый': [0.5, 0.2, 0.8]}  
          
Вывод будет выглядеть так:
Самые близкие слова к "ВАШЕ СЛОВО"
1. Первое слово
2. Второе слово
3. ...
"""
```

```python
def cosine_similarity(a, b):
    dot = sum(a_i * b_i for a_i, b_i in zip(a, b))
    prod_lens = sum(p**2 for p in a)**0.5 * sum(p**2 for p in b)**0.5

    if prod_lens == 0: print("vector was dot")

    return dot / prod_lens

def find_most_similar(target_word, words_dict, top_k=3):
        semilars = []
        try: 
            target_vec = words_dict[target_word]
        except KeyError:
             print("kew was wrong(")
             return

        for k, v in words_dict.items():
            if k == target_word: continue
            semilars.append((k, cosine_similarity(v, target_vec)))

        semilars.sort(key = lambda x: x[1], reverse = True)

        print(f"Самые близкие слова к {target_word}")

        for i, w in enumerate(semilars[:top_k]):
             print(f'{i + 1}. {w}')


        return semilars

dct1 = {'кошка': [0.9, 0.1, 0.8, 0.2],
           'собака': [0.8, 0.2, 0.7, 0.3],
           'тигр': [0.7, 0.3, 0.9, 0.1],
           'лев': [0.6, 0.4, 0.8, 0.2],
           'рыба': [0.1, 0.9, 0.2, 0.8], 
           'птица': [0.2, 0.8, 0.3, 0.7] }

find_most_similar('кошка', dct1)
```

```python
import sqlite3 as sql

def borrow_book(book_id, conn):
    conn.execute("""
        UPDATE books
        SET is_available = ?
        WHERE title = ?
    """, (False, book_id))

    conn.commit()

def return_book(book_id, conn):
    conn.execute("""
        UPDATE books
        SET is_available = ?
        WHERE title = ?
    """, (True, book_id))
    conn.commit()

def print_book(cur, book):
    cur.execute("""
        SELECT title, is_available
        FROM books
        WHERE title = ?
    """, (book,))

    print(cur.fetchone())

books = (
    ("Программирование на C# для начинающих. Основные сведения", "Алексей Николаевич Васильев", 2022, True),
    ("1984", "Джордж Оруэээ", 1948, True),
    ("Цветы для Элджернона", "Дэниел Киз", 1966, True),
    ("Грокаем алгоритмы", "Адитья Бхаргава", 2017, True),
    ("Паттерны объектно-ориентированного проектирования", "Эрих Гамма, Ричард Хелм, Ральф Джонсон, Джон Влиссидес", 1994, True)
)

conn = sql.connect('books.db')
cur = conn.cursor()

try:
    conn.execute(
    """
        CREATE TABLE IF NOT EXISTS books(
            title TEXT,
            author TEXT,
            year INTEGER,
            is_available BOOLEAN
        )
    """)

    conn.executemany("INSERT INTO books (title, author, year, is_available) VALUES (?, ?, ?, ?)", books)
    conn.commit()

    book = '1984'

    print_book(cur, book)

    borrow_book(book, conn)

    print_book(cur, book)

    return_book(book, conn)

    print_book(cur, book)

    cur.close()
    conn.execute("DROP TABLE IF EXISTS books")
except Exception as err:
    print(err.with_traceback())
finally:
    cur.close()
    conn.close()


```
