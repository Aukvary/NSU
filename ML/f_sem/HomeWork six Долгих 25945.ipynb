```python
"""Разработайте программу для ведения списка покупок. Система должна:
1. Инициализация (5 баллов)
Создать словарь покупок shopping_list:
молоко: 1
хлеб: 2
яйца: 10

2. Добавление товаров с обработкой ошибок (10 баллов)
Функция add_prod(товар, количество):
    Добавляет товар в список
    Обрабатывает ошибки:
        Отрицательное количество
        Нечисловое количество

3. Рекурсивный поиск товаров (10 баллов)
Рекурсивная функция find_prod(количество):
Ищет товары, которых нужно купить больше указанного количества
Возвращает список найденных товаров

4. Расчет стоимости (5 баллов)
Рассчитать общую стоимость на основе:
    Списка цен: [('молоко', 80), ('хлеб', 40), ('яйца', 10)]
    Вывести общую сумму
"""
```

```python
shopping_list = {
    'молоко': 1,
    'хлеб': 2,
    'яйца': 10,
}

prices = [('молоко', 80), ('хлеб', 40), ('яйца', 10)]

def get_keys(i): return list(shopping_list.keys())[i]

def add_prod(prod, count):
    try:
        if not isinstance(prod, str):
            raise TypeError('')
        shopping_list[prod] += count
    except KeyError:
        shopping_list[prod] 

def find_prod(count, idx = 0):
    if idx == len(shopping_list): return []            

    cur = shopping_list[get_keys(idx)]

    return find_prod(count, idx + 1) + ([get_keys(idx)] if cur > count else [])

def calc_price():
    s = 0
    for pair in prices:
        s += shopping_list[pair[0]] * pair[1]

    return s

add_prod('молоко', 3)

print(find_prod(2))

print(calc_price())
```
